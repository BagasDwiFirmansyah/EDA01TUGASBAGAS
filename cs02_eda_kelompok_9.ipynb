#**Nomor 1**
Apakah ada dosen yang cenderung memberikan nilai cenderung lebih tinggi (atau lebih rendah) dibandingkan dosen lainnya?

```python
nilai_dosen = nilai.groupby([nilai['NIDN']]).mean()[['nilai']]
nilai_dosen = nilai_dosen.rename({'nilai': 'mean_nilai'}, axis=1)
nilai_dosen = pd.merge(dosen, nilai_dosen, left_on='NIDN', right_index=True)[['NIDN', 'nama', 'mean_nilai']]
nilai_dosen = nilai_dosen.set_index(['NIDN'])
std = nilai.groupby([nilai['NIDN']]).std()[['nilai']]
std = std.rename({'nilai': 'std_nilai'}, axis=1)
nilai_dosen = pd.concat([nilai_dosen, std], axis=1)
nilai_dosen
```

**Interpretasi Nomor 1**

>>Berdasarkan tabel diatas didapat nilai rata-rata dan nilai standar deviasi dari keempat dosen. Berdasarkan kolom nilai rata-rata, dapat terlihat bahwa dosen bernama Pak Andri Basuki cenderung memberikan nilai yang lebih tinggi untuk para mahasiswanya. Sedangkan dosen bernama Ibu Annisa Fitria cenderung memberikan nilai yang lebih rendah untuk para mahasiswanya. Namun bila dilihat dari kolom nilai standar deviasi (simpangan baku), didapat sebaliknya bahwa Ibu Annisa Fitria cenderung memberikan nilai yang lebih tinggi sedangkan Pak Andri Basuki cenderung memberikan nilai yang lebih rendah untuk para mahasiswanya. Hal itu dikarenakan kedua dosen tersebut mengajar di jumlah kelas yang berbeda.

Sehingga dapat disimpulkan bahwa tidak ada dosen yang memberikan nilai cenderung lebih tinggi atau lebih rendah karena jumlah kelas masing-masing yang mereka ajar berbeda.

```python
Dosen = nilai.groupby(nilai['NIDN']).mean()[['nilai']]
Dosen = Dosen.rename({'nilai': 'mean_nilai'}, axis=1) 
Dosen = pd.merge(dosen, Dosen, left_on='NIDN', right_index=True)
Dosen = Dosen[['NIDN', 'nama', 'Pendidikan_Terakhir', 'mean_nilai']]
std = nilai.groupby(nilai['NIDN']).std()[['nilai']]
std = std.rename({'nilai': 'std_nilai'}, axis=1) 
Dosen = Dosen.set_index(['NIDN'])
Dosen = pd.concat([Dosen, std], axis=1)
Dosen
```

>>-Bila dilihat dari kolom nilai rata-rata, dosen dengan pendidikan terakhir S3 yaitu Pak Andri Basuki dan Pak Zulfikar cenderung memberikan nilai yang lebih tinggi bila dibandingkan dengan dosen pendidikan terakhir S2 yaitu Bu Annisa Fitria dan Pak Yudi.

Sedangkan dosen dengan pendidikan terakhir S2 yaitu Bu Annisa Fitria cenderung lebih tinggi dan dosen Pak Yudi cenderung lebih tinggi juga namun masih dibawah dosen S2 Bu Annisa Fitria dan dosen S3 Pak Zulfikar.
